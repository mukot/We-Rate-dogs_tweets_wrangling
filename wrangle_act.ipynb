## **Cleaning texts containing url**

**Define** : Remove the strings starting with https in the text column in the enhanced twitter data
    
**Code**

```python
import re
twitter_enhanced_clean['cleantext'] = twitter_enhanced_clean['text'].apply(lambda x: re.split('https:\/\/.*', str(x))[0])
```

## **Cleaning incorrect ratings**
**Define** : Cleaning the incorrectly extracted ratings, ratings with decimals and more than one ratings in the text
    
**Code**

```python
twitter_enhanced_clean.query('rating_numerator >= 15')[['cleantext','rating_numerator']]
```

```python
 # Removing more than one ratings in the text and reinserting the correct rating into the data frame
val = twitter_enhanced_clean.iloc[[313, 1165, 1202]]
new_val = val['cleantext'].str.split('/', n=1, expand=True)
new_val
```

```python
#splitting the text with the correct rating
new_val2 = new_val[1].str.split('/', n=1, expand=True)
new_val2
```

```python
#Extracting the numerator 
val["numerator"]= new_val2[0].str.extract(r'(\d+)$', expand=True).astype(float)
val = val[['tweet_id','numerator']]
val
```

```python
#rExtracting rating numerators with decimals and extracting them correctly. 
val2 = twitter_enhanced_clean.iloc[[340, 695, 763, 1712]]
new_val2 = val2['cleantext'].str.split('/', n=1, expand=True)
new_val2
```

```python
val2["numerator"]= new_val2[0].str.extract(r'(\d+\.\d+)$', expand=True).astype(float)
val2 = val2[['tweet_id','numerator']]
val2
```

```python
#Merging the new data values into the datadframe.
corrected_values = val2.append(val)
corrected_values
```

```python
#Merging the corrected values with the twitter enhanced clean data
twitter_enhanced_clean2 = pd.merge(twitter_enhanced_clean, corrected_values, on ='tweet_id', how ='left')
twitter_enhanced_clean2.info()
```

# **2. Tidyness Issues**


## Transposing all dog stages column into on single dog stage column

**Define**

Values for dog stages are given as column names which contradicts the tidyness rule.Here we transpose all the dog stages into a single column heading dog_stages.

**Code**

```python
twitter_enhanced_clean7['doggo'] = twitter_enhanced_clean7['doggo'].str.replace('None','',regex=False)
twitter_enhanced_clean7['floofer'] = twitter_enhanced_clean7['floofer'].str.replace('None','',regex=False)
twitter_enhanced_clean7['pupper'] = twitter_enhanced_clean7['pupper'].str.replace('None','',regex=False)
twitter_enhanced_clean7['puppo'] = twitter_enhanced_clean7['puppo'].str.replace('None','',regex=False) 
```

```python
twitter_enhanced_clean7['dog_stage'] = (twitter_enhanced_clean7['pupper']+' '+twitter_enhanced_clean7['doggo']+' '+twitter_enhanced_clean7['puppo']+' '+twitter_enhanced_clean7['floofer']).str.strip()
twitter_enhanced_clean7['dog_stage'] = twitter_enhanced_clean7['dog_stage'].str.replace('  ',' ',regex=False) 
twitter_enhanced_cleaned = twitter_enhanced_clean7.drop(columns=['doggo','floofer','pupper','puppo','name','rating_numerator','rating_denominator'])
```

```python
#Merging the cleaned twitter enhanced data with cleaned api data
twitter_enhanced_cleaned = pd.merge(twitter_enhanced_cleaned, api_data_clean2, on='tweet_id', how='left')
twitter_enhanced_cleaned.info()
```
